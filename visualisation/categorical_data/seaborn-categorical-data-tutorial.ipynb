### Source: [seaborn categorical data tutorial](https://seaborn.pydata.org/tutorial/categorical.html)

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set(style='ticks', color_codes=True) 
```

```python
tips = sns.load_dataset("tips")
tips
```

## Categorical scatterplots

```python
sns.catplot(x="day", y="total_bill", data=tips)
```

```python
sns.catplot(x="day", y="total_bill", jitter=False, data=tips) 
```

```python
sns.catplot(x="day", y="total_bill", kind="swarm", data=tips) 
```

```python
sns.catplot(x="day", y="total_bill", hue="sex", kind="swarm", data=tips) 
```

```python
sns.catplot(x="size", y="total_bill", kind="swarm", data=tips.query("size != 3"))
```

```python
sns.catplot(x="smoker", y="tip", order=["No", "Yes"], data=tips)
```

```python
sns.catplot(x="total_bill", y="day", hue="time", kind="swarm", data=tips)
```

```python
tips.head()
```

## Distributions of observations within categories

### Boxplots

```python
sns.catplot(x="day", y="total_bill", kind="box", data=tips) 
```

```python
sns.catplot(x="day", y="total_bill", hue="smoker", kind="box", data=tips) 
```

```python
tips["weekend"] = tips["day"].isin(['Sat', 'Sun'])
sns.catplot(x="day", y="total_bill", hue="weekend", kind="box", dodge=False, data=tips)
```

## Statistical estimation within categories
### Bar plots

```python
titanic = sns.load_dataset("titanic")
titanic.head()
```

```python
sns.catplot(x="sex", y="survived", hue="class", kind="bar", data=titanic) 
```

```python
sns.catplot(x="deck", kind="count", palette="ch:.25", data=titanic) 
```

```python
sns.catplot(y="deck", hue="class", kind="count", palette="pastel", edgecolor=".6", 
            data=titanic)
```

### Point plots

```python
sns.catplot(x="sex", y="survived", hue="class", kind="point", data=titanic) 
```

```python
sns.catplot(x="class", y="survived", hue="sex", palette={"male": "g", "female": "m"},
           markers=["^", "o"], linestyles=["-", "--"],
           kind="point", data=titanic) 
```
