```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="ticks", color_codes=True) 
sns.set_style("whitegrid")
```

```python
df_raw = pd.read_feather('../data/tmp/bulldozers-raw') 
```

```python
# prevent ellipses and show all rows/cols
pd.set_option('max_colwidth', 800)
pd.set_option('max_rows', 800)
```

```python
# which drive system is more expensive?
g = sns.catplot(x="Drive_System", y="SalePrice", kind="bar", data=df_raw) 
g.fig.set_figwidth(8.7)
g.fig.set_figheight(6.27)
g.set_xticklabels(rotation=45)
```

```python
sns.catplot(x="Stick", y="SalePrice", hue="Drive_System", kind="bar", data=df_raw)
```

```python
g = sns.catplot(x="Drive_System", y="SalePrice", kind="box", data=df_raw)
g.fig.set_figwidth(8.7)
g.fig.set_figheight(6.27)
g.set_xticklabels(rotation=45)
```
