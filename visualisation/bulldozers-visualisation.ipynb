```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="ticks", color_codes=True) 
sns.set_style("whitegrid")
```

```python
df_raw = pd.read_feather('../data/tmp/bulldozers-raw') 
```

```python
# which drive system is more expensive?
g = sns.catplot(x="Drive_System", y="SalePrice", kind="bar", data=df_raw) 
g.fig.set_figwidth(8.7)
g.fig.set_figheight(6.27)
g.set_xticklabels(rotation=45)
```

```python
sns.catplot(x="Stick", y="SalePrice", hue="Drive_System", kind="bar", data=df_raw) 
```

```python
g = sns.catplot(x="Drive_System", y="SalePrice", kind="box", data=df_raw)
g.fig.set_figwidth(8.7)
g.fig.set_figheight(6.27)
g.set_xticklabels(rotation=45)
```

```python
sns.distplot(df_raw['SalePrice'], kde=True) 
```

```python
# how does SalePrice vary with YearMade?
sns.jointplot(x="YearMade", y="SalePrice", data=df_raw) 
```

```python
# scatterplot of saleYear vs SalePrice
sns.jointplot(x="saleYear", y="SalePrice", data=df_raw) 
```

```python
# hexbin plot of saleYear vs SalePrice
with sns.axes_style("white"):
    sns.jointplot(x=df_raw["saleYear"], y=df_raw["SalePrice"], kind="hex", color="k",
                 space=0, ratio=4) 
```

```python
# KDE of saleYear and SalePrice
sns.jointplot(x="saleYear", y="SalePrice", kind="kde", data=df_raw) 
```
